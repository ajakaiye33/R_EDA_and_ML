- ** Gini index**
- **Entropy**
- **Deviance**

##### Our focus is  Entropy/Informatio gain
Entropy is originally from the field of information Theory also known as Shannon Entropy. In view of Decision Tree, Entropy is a method used to measure purity by measuring the uncertainty of a single subset
In this excercise, our goal is to predict the revenue grid(1 or 0) of the  age bands
an the formula is:
$$Entropy H(X) = -\sum p(X)\log p(X)$$

$$Information\ Gain\; I(X,Y)= H(X)-H(X|Y)$$

```python
subPop1 = dummyAgeTrain.loc[dummyAgeTrain['proc_age_band_18-21']>0.5]
subPop2 = dummyAgeTrain.loc[dummyAgeTrain['proc_age_band_18-21']<=0.5]

subPop1Index = (dummyAgeTrain['proc_age_band_18-21']>0.5).astype(int)
subPop2Index = (dummyAgeTrain['proc_age_band_18-21']<=0.5).astype(int)

print('Shape subPop1, subPop2:', sum(subPop1Index), sum(subPop2Index))

subPop1_y = y[subPop1Index]
subPop2_y = y[subPop2Index]

print('Frequency count of y for sub pop 1:', pd.value_counts(subPop1_y))
print('Frequency count of y for sub pop 2:', pd.value_counts(subPop2_y))

def entropy_index(y):
    freqCount = pd.value_counts(y)
    p0 = freqCount.loc[0]/len(y)
    p1 = freqCount.loc[1]/len(y)
    
    entropy =       -(p1*np.log(p1)) - (p0*np.log(p0))
    return entropy
print('Entropy of supop 1:', entropy_index(subPop1_y))
print('Entropy of supop 2:', entropy_index(subPop2_y))


    
def entropy_info_gain_rule(subPop1_y, subPop2_y):
    entropy1 = entropy_index(subPop1_y)
    entropy2 = entropy_index(subPop2_y)
    
    w1 = len(subPop1_y)/(len(subPop1_y) + len(subPop2_y))
    w2 = len(subPop2_y)/(len(subPop1_y) + len(subPop2_y))
    entropy_inf_gain_rule = 1. -(w1 * entropy1 + w2 * entropy2)
    return entropy_inf_gain_rule
print('entropy of rule:', entropy_info_gain_rule(subPop1_y, subPop2_y))
```

```python
def evaluateoheVar(df,target,binVar):
    print('Evaluating entropy of the OHE binary variable:',binVar)
    subPop1 = df.loc[df[binVar]>0.5]
    subPop2 = df.loc[df[binVar]<=0.5]
    
    print('Shape subPop1, subPop2:', sum(subPop1Index), sum(subPop2Index))
    subPop1_y = y[subPop1Index]
    subPop2_y = y[subPop2Index]
    
    print('Frequency count of y fir sub pop 1:',pd.value_counts(subPop1_y))
    print('Frequency count of y fir sub pop 1:',pd.value_counts(subPop2_y))
    
    print('Entropy of subpop 1:', entropy_index(subPop1_y))
    print('Entropy of subpop 2:', entropy_index(subPop2_y))
    
    entropy_of_rule = entropy_info_gain_rule(subPop1_y, subPop2_y)
    print('Entropy of rule:', entropy_of_rule)
    
```

```python
"""Applying the function: evaluateoheVar"""
df = dummyAgeTrain
target = y
binVar = 'proc_age_band_31-35'
evaluateoheVar(df, target, binVar)
```
